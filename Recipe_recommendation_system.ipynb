```python
# 사용자 입력 받기
user_id_input = input("사용자 ID를 입력하세요.: ")  # 사용자 ID 입력
user_sentence = input("오늘의 기분을 문장으로 입력하세요.: ")  # 현재 상태 입력

# OpenAI 감정 추출
response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=[
        {"role": "user", "content": f"'{user_sentence}'라는 문구를 기쁨, 분노, 불안, 슬픔, 중립 중 하나로 분류해줘"}
    ]
)
response_text = response['choices'][0]['message']['content'].strip()

#문장에서 감정만 추출
emotions = ["기쁨", "분노", "불안", "슬픔", "중립"]
emotion = next((emotion for emotion in emotions if emotion in response_text), None)

print(f"{user_id_input}의 감정: {emotion}")
```

```python
# 감정별 재료 매핑
ingredients_by_emotion = {
    "기쁨": ["과일", "아이스크림", "초콜릿", "피자", "고기", "파스타", "디저트"],
    "분노": ["녹차", "캐모마일", "베리류", "다크 초콜릿", "연어", "고등어", "고구마", "견과류", "바나나"],
    "불안": ["녹차", "요거트", "발효식품", "베리 류", "체리", "달걀", "귀리", "통곡물", "아보카도", "캐모마일"],
    "슬픔": ["다크 초콜릿", "연어", "고등어", "견과류", "바나나", "귀리", "통곡물"],
    "중립": []  # 중립은 아무 음식이나 선택 가능
}

# 감정에 해당하는 재료 리스트 가져오기
ingredients_to_search = ingredients_by_emotion.get(emotion, [])

# 감정에 따른 레시피 필터링
if emotion == "중립":
    filtered_recipes = recipe
else:
    filtered_recipes = recipe[
        recipe["분리된 재료"].apply(
            lambda x: any(ingredient in x for ingredient in ingredients_to_search)
        )
    ]
```

신규 회원일 경우 filtered_recipes에서 5개 랜덤 추천

```python
# 신규 회원 -> 바로 레시피 추출
if user_id_input not in ratings['user_id'].unique():
    final_recommendations = filtered_recipes.sample(5)
    final_recommendations['추천 점수'] = 0  # 신규 회원의 경우 기본 점수 0으로 추가
else:
    # 기존 회원 -> 사용자 기반 협업 필터링 준비
    user_item_matrix = ratings.pivot_table(index='user_id', columns='레시피 ID', values='stars').fillna(0)
    user_similarity = cosine_similarity(user_item_matrix)
    user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

    # 유사한 사용자 기반 추천
    if user_id_input in user_similarity_df.index:
      similar_users = user_similarity_df[user_id_input].sort_values(ascending=False).index[1:6]
      similar_users_ratings = user_item_matrix.loc[similar_users].mean()
      filtered_recipes['추천 점수'] = filtered_recipes['레시피 ID'].map(similar_users_ratings)

    # 빈 레시피 점수 처리
    def handle_missing_ratings(filtered_recipes, user_item_matrix, default_rating=3):
      missing_recipes = filtered_recipes[~filtered_recipes['레시피 ID'].isin(user_item_matrix.columns)]
      missing_recipes['추천 점수'] = default_rating
      filtered_recipes = pd.concat([filtered_recipes, missing_recipes])
      return filtered_recipes

    filtered_recipes = handle_missing_ratings(filtered_recipes, user_item_matrix)

    # 추천된 레시피 상위 5개 정보 가져오기
    top_5_recipes = filtered_recipes[['레시피 ID', '추천 점수']].sort_values(by='추천 점수', ascending=False).head(5)
    recommended_recipe_ids = top_5_recipes['레시피 ID'].tolist()

    # 상위 5개 레시피 정보를 원본 데이터에서 추출
    final_recommendations = recipe[recipe['레시피 ID'].isin(recommended_recipe_ids)].copy()
    final_recommendations = pd.merge(
        final_recommendations,
        top_5_recipes,
        on='레시피 ID'
        )

    # 원하는 열만 선택
    final_recommendations = final_recommendations[
        ['레시피 ID', '요리제목', '재료', '조리순서', '분리된 재료', '추천 점수']
    ].sort_values(by='추천 점수', ascending=False)
```

```python
# 결과 출력
print(f"\n{user_id_input} 사용자 맞춤 추천 레시피:")
print(final_recommendations)
final_recommendations
```

```python
# OpenAI를 이용해 한국어를 영어로 번역
def translate_to_english(recipe_name):
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[
            {"role": "user", "content": f"Translate the following dish name to English in a safe and neutral manner, do not use 'slut': '{recipe_name}'."}
        ]
    )
    english_name = response['choices'][0]['message']['content'].strip()
    return english_name

# OpenAI를 이용해 음식 이미지 생성
def generate_food_image(recipe_name):
    prompt = (
        f"A delicious and realistic image of {recipe_name}, presented beautifully on a white plate. "
        "The setting is clean and professional, perfect for a cookbook."
    )
    response = openai.Image.create(
        prompt=prompt,
        n=1,
        size="256x256"
    )
    image_url = response['data'][0]['url']
    return image_url

# 이미지 생성 및 출력
for index, row in final_recommendations.iterrows():
    recipe_name_korean = row['요리제목']
    print(f"\nTranslating '{recipe_name_korean}' to English...")

    # 요리 제목 영어로 변환
    recipe_name_english = translate_to_english(recipe_name_korean)
    print(f"Translated name: {recipe_name_english}")

    print(f"\nGenerating image for {recipe_name_english}...")

    # 이미지 생성
    image_url = generate_food_image(recipe_name_english)
    print(f"Image for {recipe_name_korean} ({recipe_name_english}): {image_url}")
    display(Image(url=image_url))
```

```python
import requests
from bs4 import BeautifulSoup
from IPython.display import display, HTML

# 상위 5개 추천 레시피 정보
top_5_recipes = final_recommendations[['레시피 ID', '요리제목', '추천 점수']]


def search_recipe_10000recipe(query):
    """만개의레시피에서 음식 이름으로 검색하고 첫 번째 결과 링크 반환"""
    base_url = "https://www.10000recipe.com"
    search_url = f"{base_url}/recipe/list.html?q={query}"
    headers = {"User-Agent": "Mozilla/5.0"}

    response = requests.get(search_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        recipe_link = soup.select_one(".common_sp_list_ul .common_sp_thumb a")
        if recipe_link:
            recipe_url = base_url + recipe_link['href']
            return recipe_url
    return None

def get_recipe_details(recipe_url):
    """레시피 상세 페이지에서 제목, 재료, 조리 단계를 추출"""
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(recipe_url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        title = soup.select_one(".view2_summary h3").text.strip()
        ingredients = [li.text.strip() for li in soup.select(".ready_ingre3 ul li")]
        steps = [step.text.strip() for step in soup.select(".view_step .media-body")]
        return {
            "title": title,
            "ingredients": ingredients,
            "steps": steps,
            "url": recipe_url
        }
    return None

# 추천된 각 레시피에 대해 상세 정보 가져오기
recipe_details_list = []

for index, row in top_5_recipes.iterrows():
    recipe_title = row['요리제목']
    print(f"Searching for '{recipe_title}' on 만개의레시피...")
    recipe_url = search_recipe_10000recipe(recipe_title)

    if recipe_url:
        recipe_details = get_recipe_details(recipe_url)
        if recipe_details:
            recipe_details['추천 점수'] = row['추천 점수']  # 점수 추가
            recipe_details_list.append(recipe_details)
            print(f"레시피 찾음: {recipe_details['title']} ({recipe_details['url']})")
    else:
        print(f"'{recipe_title}'에 해당하는 레시피를 찾을 수 없습니다.")

# 결과 출력 (HTML 형식으로 보기 좋게 출력)
for recipe in recipe_details_list:
    display(HTML(f"""
        <h3>{recipe['title']}</h3>
        <p><b>추천 점수:</b> {recipe['추천 점수']}</p>
        <p><b>링크:</b> <a href="{recipe['url']}" target="_blank">{recipe['url']}</a></p>
        <p><b>재료:</b> {', '.join(recipe['ingredients'])}</p>
        <p><b>조리 단계:</b></p>
        <ol>{"".join(f"<li>{step}</li>" for step in recipe['steps'])}</ol>
        <hr>
    """))
```
